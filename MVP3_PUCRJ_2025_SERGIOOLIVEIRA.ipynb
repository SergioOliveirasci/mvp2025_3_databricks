Nesta fase, **n√£o s√£o aplicadas transforma√ß√µes ou tratamentos anal√≠ticos**, garantindo a rastreabilidade e a possibilidade de reprocessamento dos dados, caso necess√°rio.

Os dados utilizados neste projeto j√° se encontram carregados no ambiente Databricks e s√£o acessados por meio de uma **tabela Delta**, criada a partir do dataset original. Essa abordagem permite:

O c√≥digo a seguir realiza a leitura da tabela correspondente aos dados brutos e exibe seu conte√∫do inicial, permitindo a valida√ß√£o da estrutura, dos tipos de dados e das colunas dispon√≠veis antes do in√≠cio das transforma√ß√µes subsequentes.

## 4. Transforma√ß√£o dos Dados (Camada Silver)

A Camada Silver representa a etapa intermedi√°ria do pipeline de dados, na qual os dados brutos provenientes da Camada Bronze passam por **processos de limpeza, padroniza√ß√£o e organiza√ß√£o**, com o objetivo de torn√°-los **consistentes, confi√°veis e adequados para an√°lises anal√≠ticas**.


Nesta fase, s√£o aplicadas transforma√ß√µes controladas que **n√£o alteram o significado dos dados**, mas corrigem problemas comuns encontrados em bases reais, como inconsist√™ncias de formata√ß√£o, varia√ß√µes de texto e nomenclaturas inadequadas para an√°lise.

As principais atividades realizadas nesta etapa incluem:
- **Padroniza√ß√£o de campos textuais**, por meio da remo√ß√£o de espa√ßos em branco e normaliza√ß√£o de letras (ex.: uso de min√∫sculas);
- **Renomea√ß√£o de colunas**, adequando os nomes para um padr√£o t√©cnico consistente, facilitando consultas SQL e manipula√ß√£o anal√≠tica;
- **Prepara√ß√£o dos dados para an√°lise**, garantindo que as colunas estejam semanticamente claras e prontas para agrega√ß√µes, filtros e m√©tricas posteriores.

Essa camada desempenha um papel fundamental na **garantia da qualidade dos dados**, servindo como base confi√°vel para a constru√ß√£o da Camada Gold, onde ser√£o realizadas as an√°lises finais e a gera√ß√£o de insights.

Ao final deste processo, os dados permanecem com a mesma granularidade do dataset original, por√©m organizados de forma padronizada, assegurando **rastreabilidade, reprodutibilidade e clareza anal√≠tica** ao pipeline desenvolvido.


```python
from pyspark.sql.functions import col, lower, trim, when
from pyspark.sql.types import DoubleType

df_silver = (
    df_bronze
    # Limpeza/padroniza√ß√£o de texto (mantendo significado)
    .withColumn("Medicine Name", trim(col("Medicine Name")))
    .withColumn("Manufacturer", trim(col("Manufacturer")))
    .withColumn("Composition", trim(col("Composition")))
    .withColumn("Uses", when(col("Uses").isNotNull(), lower(trim(col("Uses")))).otherwise(col("Uses")))
    .withColumn("Side_effects", when(col("Side_effects").isNotNull(), lower(trim(col("Side_effects")))).otherwise(col("Side_effects")))
    
    # Renomea√ß√£o padronizada para an√°lise e SQL
    .withColumnRenamed("Medicine Name", "Medicine_Name")
    .withColumnRenamed("Image URL", "Image_URL")
    .withColumnRenamed("Excellent Review %", "Excellent_Review")
    .withColumnRenamed("Average Review %", "Average_Review")
    .withColumnRenamed("Poor Review %", "Poor_Review")

    # Garantia de tipos num√©ricos para m√©tricas (caso tenham vindo como string)
    .withColumn("Excellent_Review", col("Excellent_Review").cast(DoubleType()))
    .withColumn("Average_Review", col("Average_Review").cast(DoubleType()))
    .withColumn("Poor_Review", col("Poor_Review").cast(DoubleType()))
)

(
    df_silver.write
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("workspace.default.medicine_details_silver")
)

display(df_silver)
print("Linhas Bronze:", df_bronze.count())
print("Linhas Silver:", df_silver.count())
```

### Valida√ß√£o P√≥s-Transforma√ß√£o (Camada Silver)

A execu√ß√£o da etapa Silver confirma que os dados passaram por **padroniza√ß√£o e limpeza b√°sica** com sucesso, mantendo a granularidade do dataset original.  
Observa-se que:

- Os campos textuais relevantes foram **normalizados** (remo√ß√£o de espa√ßos excedentes e padroniza√ß√£o de caixa em `Uses` e `Side_effects`), reduzindo inconsist√™ncias de escrita e facilitando an√°lises por agrupamento.
- As colunas foram **renomeadas para um padr√£o consistente** (snake_case), melhorando a legibilidade e simplificando consultas SQL.
- Os percentuais de avalia√ß√£o (`Excellent_Review`, `Average_Review`, `Poor_Review`) foram preparados para uso anal√≠tico, permitindo agrega√ß√µes e compara√ß√µes de forma confi√°vel.

A Camada Silver passa a atuar como a **base confi√°vel** para a etapa seguinte (Camada Gold), onde ser√£o realizadas agrega√ß√µes, m√©tricas e consultas orientadas √†s perguntas de neg√≥cio.


**Confer√™ncia de integridade:** foi realizada a compara√ß√£o de quantidade de registros entre Bronze e Silver para garantir que as transforma√ß√µes n√£o causaram perdas indevidas de dados.


## 5. Camada Anal√≠tica (Gold)

A Camada Gold representa a etapa final do pipeline, na qual os dados j√° tratados e padronizados na Camada Silver s√£o **agregados e estruturados** para suportar an√°lises anal√≠ticas e responder diretamente √†s **perguntas de neg√≥cio** definidas no in√≠cio do projeto.

Nesta camada, s√£o criadas tabelas orientadas a consumo anal√≠tico (por exemplo, **m√©tricas por fabricante, uso terap√™utico ou medicamento**), com o objetivo de:
- Reduzir a complexidade das consultas finais (deixando-as mais simples e rastre√°veis);
- Melhorar desempenho na explora√ß√£o anal√≠tica (dados j√° consolidados);
- Garantir consist√™ncia das m√©tricas, evitando m√∫ltiplas vers√µes de um mesmo c√°lculo.

Neste primeiro artefato da Camada Gold, √© constru√≠da uma tabela agregada **por fabricante**, calculando m√©tricas de avalia√ß√£o m√©dia dos usu√°rios, como:
- m√©dia de avalia√ß√µes excelentes (*Excellent Review*);
- m√©dia de avalia√ß√µes m√©dias (*Average Review*).

O resultado √© persistido como tabela Delta na camada Gold e utilizado nas an√°lises subsequentes.


```python
from pyspark.sql.functions import avg, round, col

df_gold = (
    spark.table("workspace.default.medicine_details_silver")
    .groupBy("Manufacturer")
    .agg(
        round(avg("Excellent_Review"), 2).alias("avg_excellent_review"),
        round(avg("Average_Review"), 2).alias("avg_average_review"),
        round(avg("Poor_Review"), 2).alias("avg_poor_review")
    )
    .orderBy(col("avg_excellent_review").desc())
)

(
    df_gold.write
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("workspace.default.medicine_manufacturer_gold")
)

display(df_gold)
```

## 6. An√°lise da Qualidade dos Dados

Nesta etapa √© realizada a an√°lise da qualidade dos dados consolidados na **Camada Silver**, com o objetivo de garantir que os atributos estejam consistentes, confi√°veis e adequados para a constru√ß√£o das an√°lises anal√≠ticas na Camada Gold.

A verifica√ß√£o de qualidade vai al√©m da simples identifica√ß√£o de valores nulos, contemplando tamb√©m regras b√°sicas de consist√™ncia e coer√™ncia dos dados, tais como:
- presen√ßa de valores nulos e campos textuais vazios;
- verifica√ß√£o de faixas v√°lidas para os percentuais de avalia√ß√£o dos usu√°rios;
- valida√ß√£o da consist√™ncia da soma dos percentuais de avalia√ß√£o;
- identifica√ß√£o de poss√≠veis registros duplicados com base em uma chave l√≥gica.

Essas an√°lises fornecem evid√™ncias de controle de qualidade ao pipeline e aumentam a confiabilidade das m√©tricas e insights produzidos nas etapas seguintes do projeto.

```python
from pyspark.sql import functions as F

df_quality = spark.table("workspace.default.medicine_details_silver")

# =========================
# 1) Valores Nulos por Coluna
# =========================
nulls_df = df_quality.select([
    F.sum(F.col(c).isNull().cast("int")).alias(c)
    for c in df_quality.columns
])

display(nulls_df)

# =========================
# 2) Campos Vazios (strings)
# =========================
string_cols = [name for name, dtype in df_quality.dtypes if dtype == "string"]

empties_df = df_quality.select([
    F.sum((F.trim(F.col(c)) == "").cast("int")).alias(c)
    for c in string_cols
])

display(empties_df)

# =========================
# 3) Percentuais fora do intervalo [0, 100]
# =========================
pct_cols = ["Excellent_Review", "Average_Review", "Poor_Review"]

out_of_range_df = df_quality.select(
    *[
        F.sum((F.col(c) < 0).cast("int")).alias(f"{c}_below_0")
        for c in pct_cols
    ],
    *[
        F.sum((F.col(c) > 100).cast("int")).alias(f"{c}_above_100")
        for c in pct_cols
    ]
)

display(out_of_range_df)

# =========================
# 4) Consist√™ncia da soma dos percentuais (~100)
#    Toler√¢ncia: 99 a 101
# =========================
df_sum_check = df_quality.withColumn(
    "review_sum",
    F.col("Excellent_Review") + F.col("Average_Review") + F.col("Poor_Review")
)

total_rows = df_quality.count()
inconsistent_sum = df_sum_check.filter(
    (F.col("review_sum") < 99) | (F.col("review_sum") > 101)
).count()

display(
    spark.createDataFrame(
        [(total_rows, inconsistent_sum)],
        ["total_registros", "registros_com_soma_inconsistente"]
    )
)

# =========================
# 5) Duplicidade por chave l√≥gica
#    (Medicamento + Fabricante)
# =========================
duplicates_df = (
    df_quality
    .groupBy("Medicine_Name", "Manufacturer")
    .count()
    .filter(F.col("count") > 1)
    .orderBy(F.col("count").desc())
)

display(duplicates_df)
```

A an√°lise de qualidade realizada permite avaliar se os dados est√£o aptos para consumo anal√≠tico na Camada Gold.

### 7.1 Pergunta 1 ‚Äî Quais fabricantes possuem os medicamentos com melhores avalia√ß√µes dos usu√°rios?

**Objetivo:** identificar fabricantes com maior aceita√ß√£o m√©dia, utilizando como m√©trica o percentual m√©dio de avalia√ß√µes excelentes (`Excellent_Review`) e, como apoio, o percentual m√©dio de avalia√ß√µes m√©dias (`Average_Review`), com base na tabela agregada da Camada Gold.

```python
# 7.1 - Ranking de fabricantes por m√©dia de avalia√ß√µes excelentes (Camada Gold)

from pyspark.sql import functions as F

df_manu_gold = spark.table("workspace.default.medicine_manufacturer_gold")

df_q1 = (
    df_manu_gold
    .select(
        "Manufacturer",
        F.col("avg_excellent_review"),
        F.col("avg_average_review")
    )
    .orderBy(F.col("avg_excellent_review").desc())
)

display(df_q1)
```

**Coment√°rio ap√≥s o resultado (7.1):**  
Os resultados indicam diversos fabricantes com m√©dia de avalia√ß√µes excelentes igual a 100%. Esse comportamento sugere que, para esses fabricantes, os medicamentos dispon√≠veis na base receberam exclusivamente avalia√ß√µes classificadas como excelentes.

Entretanto, √© importante destacar que esse resultado pode estar associado a **baixo n√∫mero de medicamentos ou avalia√ß√µes por fabricante**, o que limita a compara√ß√£o direta entre eles. Assim, a an√°lise deve ser interpretada de forma descritiva, indicando percep√ß√£o positiva dos usu√°rios, mas sem permitir infer√™ncia conclusiva sobre superioridade absoluta entre fabricantes.

### 7.2 Pergunta 2 ‚Äî Existe rela√ß√£o entre o uso terap√™utico e a avalia√ß√£o positiva dos medicamentos?

**Objetivo:** avaliar se determinadas categorias de **uso terap√™utico** (`Uses`) est√£o associadas a maiores m√©dias de avalia√ß√µes excelentes.  
Como `Uses` √© um campo textual (podendo conter mais de um uso), nesta an√°lise ser√° utilizada uma abordagem por texto padronizado presente na Camada Silver.

```python
# 7.2 - M√©dia de avalia√ß√µes por "Uses" (Camada Silver)
# Observa√ß√£o: esta an√°lise usa Silver porque o campo "Uses" √© textual e pode exigir padroniza√ß√£o.

from pyspark.sql import functions as F

df_silver = spark.table("workspace.default.medicine_details_silver")

df_q2 = (
    df_silver
    .groupBy("Uses")
    .agg(
        F.round(F.avg("Excellent_Review"), 2).alias("avg_excellent_review"),
        F.count("*").alias("qtd_registros")
    )
    .orderBy(F.col("avg_excellent_review").desc(), F.col("qtd_registros").desc())
)

display(df_q2)
```

**Coment√°rio ap√≥s o resultado (7.2):**  
A an√°lise mostra que diversos usos terap√™uticos apresentam m√©dia de avalia√ß√µes excelentes igual a 100%. Esse comportamento sugere alta aceita√ß√£o dos medicamentos associados a esses usos.

No entanto, o campo `Uses` √© textual e pode representar combina√ß√µes de m√∫ltas indica√ß√µes, o que gera grande granularidade e poss√≠veis amostras pequenas. Por esse motivo, a vari√°vel `qtd_registros` deve ser considerada na interpreta√ß√£o, pois usos com poucos registros podem apresentar m√©dias extremas sem representar uma tend√™ncia geral do conjunto de dados.

### 7.3 Pergunta 3 ‚Äî Medicamentos com mais efeitos colaterais tendem a receber piores avalia√ß√µes?

**Objetivo:** criar uma m√©trica derivada de quantidade de efeitos colaterais com base no texto de `Side_effects` e comparar a m√©dia de avalia√ß√µes excelentes conforme essa quantidade aumenta.

```python
# 7.3 - Rela√ß√£o entre quantidade de efeitos colaterais e avalia√ß√µes
from pyspark.sql import functions as F

df_silver = spark.table("workspace.default.medicine_details_silver")

# M√©trica simples: n√∫mero aproximado de efeitos colaterais = quantidade de itens separados por v√≠rgula
df_side = df_silver.withColumn(
    "num_side_effects",
    F.when(F.col("Side_effects").isNull() | (F.trim(F.col("Side_effects")) == ""), F.lit(0))
     .otherwise(F.size(F.split(F.col("Side_effects"), ",")))
)

df_q3 = (
    df_side
    .groupBy("num_side_effects")
    .agg(
        F.round(F.avg("Excellent_Review"), 2).alias("avg_excellent_review"),
        F.round(F.avg("Poor_Review"), 2).alias("avg_poor_review"),
        F.count("*").alias("qtd_registros")
    )
    .orderBy(F.col("num_side_effects").asc())
)

display(df_q3)
```

**Coment√°rio ap√≥s o resultado (7.3):**  
O resultado indica que a maioria dos registros apresenta uma √∫nica faixa de quantidade de efeitos colaterais, o que impede a compara√ß√£o entre diferentes n√≠veis de efeitos adversos.

Dessa forma, nesta base de dados, n√£o foi poss√≠vel identificar uma rela√ß√£o clara entre o aumento do n√∫mero de efeitos colaterais e a varia√ß√£o das avalia√ß√µes. A an√°lise permanece descritiva e evidencia uma limita√ß√£o do dataset para esse tipo de infer√™ncia, que poderia ser explorada em trabalhos futuros com maior diversidade de registros.

### 7.4 Pergunta 4 ‚Äî Quais medicamentos apresentam maior percentual de avalia√ß√µes excelentes?

**Objetivo:** identificar medicamentos com maior valor de `Excellent_Review` e apresentar um ranking com base na Camada Silver.

```python
# 7.4 - Ranking de medicamentos por Excellent_Review (Camada Silver)

from pyspark.sql import functions as F

df_silver = spark.table("workspace.default.medicine_details_silver")

df_q4 = (
    df_silver
    .select(
        "Medicine_Name",
        "Manufacturer",
        "Excellent_Review",
        "Average_Review",
        "Poor_Review"
    )
    .orderBy(F.col("Excellent_Review").desc())
)

display(df_q4)
```

**Coment√°rio ap√≥s o resultado (7.4):**  
O ranking apresenta diversos medicamentos com 100% de avalia√ß√µes classificadas como excelentes. Esse resultado indica percep√ß√£o extremamente positiva dos usu√°rios para esses produtos espec√≠ficos.

Entretanto, a aus√™ncia de avalia√ß√µes m√©dias ou ruins sugere que esses medicamentos podem possuir n√∫mero reduzido de avalia√ß√µes ou perfil de uso espec√≠fico. Assim, o ranking deve ser interpretado como indicativo de alta aceita√ß√£o, mas n√£o como uma compara√ß√£o definitiva entre todos os medicamentos dispon√≠veis na base.

### 7.5 Pergunta 5 ‚Äî H√° concentra√ß√£o de avalia√ß√µes positivas em poucos fabricantes?

**Objetivo:** avaliar se as avalia√ß√µes excelentes est√£o concentradas em poucos fabricantes, calculando a participa√ß√£o dos Top 5 fabricantes (ordenados por m√©dia de avalia√ß√µes excelentes) sobre o total de fabricantes.

```python
# 7.5 - Concentra√ß√£o de avalia√ß√µes positivas (Top N fabricantes) - Camada Gold

from pyspark.sql import functions as F

df_manu_gold = spark.table("workspace.default.medicine_manufacturer_gold")

# Top 5 fabricantes por avg_excellent_review
top_n = 5

df_top = (
    df_manu_gold
    .orderBy(F.col("avg_excellent_review").desc())
    .limit(top_n)
)

# Somat√≥rios (m√©trica simples de concentra√ß√£o com base nas m√©dias)
sum_top = df_top.agg(F.sum("avg_excellent_review").alias("sum_top")).collect()[0]["sum_top"]
sum_all = df_manu_gold.agg(F.sum("avg_excellent_review").alias("sum_all")).collect()[0]["sum_all"]

share_top = (sum_top / sum_all) * 100 if sum_all else None

display(df_top)

print(f"Participa√ß√£o aproximada do Top {top_n} (com base na soma das m√©dias de Excellent_Review): {share_top:.2f}%")
```

**Coment√°rio ap√≥s o resultado (7.5):**  
A an√°lise do Top 5 fabricantes indica que, embora esses fabricantes apresentem m√©dias m√°ximas de avalia√ß√µes excelentes, a participa√ß√£o conjunta representa aproximadamente 1,59% do total considerado.

Esse resultado sugere **baixa concentra√ß√£o de avalia√ß√µes positivas**, indicando que a percep√ß√£o favor√°vel dos usu√°rios est√° distribu√≠da entre diversos fabricantes, e n√£o concentrada em poucos grupos dominantes. Ressalta-se que a m√©trica utilizada √© aproximada, baseada na soma das m√©dias, e an√°lises futuras poderiam ponderar pelo volume de medicamentos ou avalia√ß√µes por fabricante.
